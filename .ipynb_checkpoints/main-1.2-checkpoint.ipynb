```python
# Cellule 1 : Imports
import numpy as np
import random
import math
import copy
import time
from IPython.display import clear_output, HTML, display, update_display
from tqdm.notebook import tqdm
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D 
import seaborn as sns
import os
import csv
from numba import njit
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import matplotlib

# Fixer la graine pour la reproductibilité
random.seed(56)
np.random.seed(56)

sns.set_theme(style="whitegrid", context="notebook")
plt.rcParams['figure.dpi'] = 100
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
plt.rcParams['legend.fontsize'] = 10
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['font.family'] = 'sans-serif'
```

```python
# Cellule 2 : Fonctions utilitaires et Fonction de Coût
def generate_sudoku(n, holes):
    s = math.isqrt(n)
    def pattern(r, c): return (s * (r % s) + r // s + c) % n
    def shuffle(amt): return random.sample(amt, len(amt))
    r_base = range(s)
    rows = [g * s + r for g in shuffle(r_base) for r in shuffle(r_base)]
    cols = [g * s + c for g in shuffle(r_base) for c in shuffle(r_base)]
    nums = shuffle(range(1, n + 1))
    board = [[nums[pattern(r, c)] for c in cols] for r in rows]
    for _ in range(holes):
        board[random.randrange(n)][random.randrange(n)] = 0
    return [val for row in board for val in row]

def get_initial_board(input_string):
    if isinstance(input_string, str):
        data = [int(c) for c in input_string]
    else:
        data = input_string
    n = int(len(data)**0.5)
    return np.array(data, dtype=np.int16).reshape((n, n))

def initialize_grid(fixed_board):
    n = len(fixed_board)
    s = math.isqrt(n)
    current_board = np.copy(fixed_board)
    fixed_mask = (fixed_board != 0) 
    
    for i in range(0, n, s):
        for j in range(0, n, s):
            block = current_board[i:i+s, j:j+s]
            existing = set(block[block != 0])
            missing = [n_val for n_val in range(1, n + 1) if n_val not in existing]
            random.shuffle(missing)
            
            idx = 0
            for r in range(i, i+s):
                for c in range(j, j+s):
                    if current_board[r, c] == 0:
                        current_board[r, c] = missing[idx]
                        idx += 1
                        
    return current_board, fixed_mask

@njit(fastmath=True)
def calculate_energy(board):
    n = len(board)
    score = 0
    for i in range(n):
        unique_row = set()
        for x in board[i, :]:
            unique_row.add(x)
        score += (n - len(unique_row))
        
    for j in range(n):
        unique_col = set()
        for x in board[:, j]:
            unique_col.add(x)
        score += (n - len(unique_col))
    return score

def format_duration(seconds):
    days = int(seconds // 86400)
    seconds %= 86400
    hours = int(seconds // 3600)
    seconds %= 3600
    minutes = int(seconds // 60)
    seconds %= 60
    
    parts = []
    if days > 0:
        parts.append(f"{days}j")
    if hours > 0:
        parts.append(f"{hours}h")
    if minutes > 0:
        parts.append(f"{minutes}min")
    
    if days == 0 and hours == 0 and minutes == 0:
        if seconds == 0:
            parts.append("0s")
        elif seconds < 10:
            parts.append(f"{seconds:.2f}s".rstrip('0').rstrip('.'))
        else:
            parts.append(f"{int(seconds)}s")
    else:
        parts.append(f"{int(seconds)}s")
        
    return "".join(parts)

def display_sudoku(grid):
    n = len(grid)
    s = math.isqrt(n)
    conflicts = np.zeros((n, n), dtype=bool)
    for i in range(n):
        row = grid[i, :]
        vals, counts = np.unique(row, return_counts=True)
        dupes = vals[(vals != 0) & (counts > 1)]
        conflicts[i, np.isin(row, dupes)] = True
        
        col = grid[:, i]
        vals, counts = np.unique(col, return_counts=True)
        dupes = vals[(vals != 0) & (counts > 1)]
        conflicts[np.isin(col, dupes), i] = True

    for r0 in range(0, n, s):
        for c0 in range(0, n, s):
            block = grid[r0:r0+s, c0:c0+s]
            vals, counts = np.unique(block, return_counts=True)
            dupes = vals[(vals != 0) & (counts > 1)]
            if len(dupes) > 0:
                conflicts[r0:r0+s, c0:c0+s] |= np.isin(block, dupes)
    
    cell_size = max(20, min(40, 800 // n))
    table_w = n * cell_size + (n // s) * 2
    
    html = [f'<div style="display:flex; flex-direction:column; align-items:center; margin:20px auto;">']
    html.append(f'<table style="table-layout:fixed; width:{table_w}px; border:3px solid #000; border-collapse:collapse; background-color:#fff;">')
    
    for r in range(n):
        b_bottom = "3px solid #000" if (r + 1) % s == 0 and r != n - 1 else "1px solid #ddd"
        html.append(f'<tr style="height:{cell_size}px; border-bottom:{b_bottom};">')
        for c in range(n):
            b_right = "3px solid #000" if (c + 1) % s == 0 and c != n - 1 else "1px solid #ddd"
            if conflicts[r, c]:
                bg_color, color = "#fee2e2", "#dc2626"
            else:
                bg_color = "#ffffff" if ((r // s) + (c // s)) % 2 == 0 else "#f9fafb"
                color = "#1f2937"
            
            val = int(grid[r][c])
            val_str = str(val) if val != 0 else ""
            html.append(f'<td style="width:{cell_size}px; height:{cell_size}px; padding:0; text-align:center; vertical-align:middle; '
                        f'border-right:{b_right}; background-color:{bg_color}; color:{color}; font-family:monospace; '
                        f'font-weight:bold; font-size:{cell_size//2}px; line-height:{cell_size}px;">{val_str}</td>')
        html.append('</tr>')
    
    html.append('</table>')
    energy = calculate_energy(grid)
    status = f"Énergie : {energy} {'(Résolu)' if energy == 0 else '(En cours)'}"
    html.append(f'<div style="margin-top:10px; font-family:sans-serif; font-size:14px; font-weight:bold; color:{"#059669" if energy==0 else "#4b5563"}">{status}</div></div>')
    display(HTML("".join(html)))
```

```python
# Cellule 3 : Le Solveur Monte Carlo
@njit(fastmath=True)
def run_batch(board, block_coords, block_counts, 
              current_energy, best_energy, best_board,
              row_counts, col_counts,
              temp, cooling_rate, initial_temp, 
              steps_batch, total_steps_done, 
              n_accepted, n_rejected, n_improved, n_restarts):
    
    n = len(board)
    num_blocks = len(block_counts)
    local_step = 0
    
    if current_energy == 0:
        return current_energy, best_energy, temp, total_steps_done, n_accepted, n_rejected, n_improved, n_restarts

    while local_step < steps_batch:
        if current_energy == 0:
            break

        rand_blk_idx = np.random.randint(0, num_blocks)
        cnt = block_counts[rand_blk_idx]
        
        if cnt < 2:
            local_step += 1
            continue
            
        # Optimisation : Sélection garantie de deux indices différents sans boucle while
        idx1 = np.random.randint(0, cnt)
        idx2_offset = np.random.randint(1, cnt)
        idx2 = (idx1 + idx2_offset) % cnt
            
        r1, c1 = block_coords[rand_blk_idx, idx1]
        r2, c2 = block_coords[rand_blk_idx, idx2]
        
        v1 = board[r1, c1]
        v2 = board[r2, c2]
        
        if v1 == v2:
            n_rejected += 1
            local_step += 1
            continue
            
        d_e = 0
        if r1 != r2:
            if row_counts[r1, v1] == 1: d_e += 1
            if row_counts[r1, v2] == 0: d_e -= 1
            if row_counts[r2, v2] == 1: d_e += 1
            if row_counts[r2, v1] == 0: d_e -= 1
            
        if c1 != c2:
            if col_counts[c1, v1] == 1: d_e += 1
            if col_counts[c1, v2] == 0: d_e -= 1
            if col_counts[c2, v2] == 1: d_e += 1
            if col_counts[c2, v1] == 0: d_e -= 1

        if d_e <= 0 or (temp > 1e-9 and np.random.random() < math.exp(-d_e / temp)):
            if d_e < 0: n_improved += 1
            n_accepted += 1
            
            board[r1, c1] = v2
            board[r2, c2] = v1
            
            row_counts[r1, v1] -= 1; row_counts[r1, v2] += 1
            row_counts[r2, v2] -= 1; row_counts[r2, v1] += 1
            col_counts[c1, v1] -= 1; col_counts[c1, v2] += 1
            col_counts[c2, v2] -= 1; col_counts[c2][v1] += 1
            
            current_energy += d_e
            
            if current_energy < best_energy:
                best_energy = current_energy
                for rr in range(n):
                    for cc in range(n):
                        best_board[rr, cc] = board[rr, cc]
        else:
            n_rejected += 1
            
        temp *= cooling_rate
        
        if temp < 0.005 and current_energy > 0:
            n_restarts += 1
            temp = initial_temp
            current_energy = best_energy
            for rr in range(n):
                for cc in range(n):
                    board[rr, cc] = best_board[rr, cc]
            
            row_counts[:] = 0
            col_counts[:] = 0
            for rr in range(n):
                for cc in range(n):
                    val = board[rr, cc]
                    row_counts[rr, val] += 1
                    col_counts[cc, val] += 1
        
        local_step += 1
        total_steps_done += 1
        
    return current_energy, best_energy, temp, total_steps_done, n_accepted, n_rejected, n_improved, n_restarts


def solve_sudoku_annealing(initial_grid_str, max_steps=None, initial_temp=1.0, cooling_rate=None, infinite_mode=False, 
                           pbar_iter=None, pbar_energy=None):
    t_start = time.time()
    fixed_board = get_initial_board(initial_grid_str)
    n = len(fixed_board)
    s = int(math.isqrt(n))
    
    if infinite_mode:
        if cooling_rate is None: cooling_rate = 0.99995
        steps_limit = 1000000000
    else:
        if max_steps is None: max_steps = int((n**4) * s * 1.5)
        steps_limit = int(max_steps) # Force int conversion for tqdm safety
        if cooling_rate is None: cooling_rate = 1.0 - (6.0 / max_steps)
        
    current_board, fixed_mask = initialize_grid(fixed_board)
    
    max_cells_per_block = s * s
    num_blocks = (n // s) ** 2
    block_coords_np = np.zeros((num_blocks, max_cells_per_block, 2), dtype=np.int16)
    block_counts_np = np.zeros(num_blocks, dtype=np.int16)
    
    blk_idx = 0
    for r0 in range(0, n, s):
        for c0 in range(0, n, s):
            count = 0
            for r in range(r0, r0+s):
                for c in range(c0, c0+s):
                    if not fixed_mask[r, c]:
                        block_coords_np[blk_idx, count, 0] = r
                        block_coords_np[blk_idx, count, 1] = c
                        count += 1
            block_counts_np[blk_idx] = count
            blk_idx += 1

    current_energy = calculate_energy(current_board)
    initial_energy_val = current_energy
    best_energy = current_energy
    best_board = current_board.copy()
    temp = initial_temp
    
    row_counts = np.zeros((n, n + 1), dtype=np.int16)
    col_counts = np.zeros((n, n + 1), dtype=np.int16)
    for r in range(n):
        for c in range(n):
            val = current_board[r, c]
            row_counts[r, val] += 1
            col_counts[c, val] += 1

    n_accepted = 0
    n_rejected = 0
    n_improved = 0
    n_restarts = 0
    total_steps = 0
    
    own_pbars = False
    if pbar_iter is None:
        own_pbars = True
        # Ajout de mininterval pour éviter le flood si exécuté seul
        pbar_iter = tqdm(total=steps_limit, desc=f"Optimisation {n}x{n}", leave=False, position=1, colour='blue', mininterval=0.2)
    else:
        pbar_iter.reset(total=steps_limit)
        pbar_iter.set_description(f"Optimisation {n}x{n}")
        
    if pbar_energy is None:
        if own_pbars: 
            pbar_energy = tqdm(total=initial_energy_val, desc="Convergence (Énergie)", leave=False, position=2, 
                               bar_format='{l_bar}{bar}| {n_fmt}/{total_fmt} [Err: {n_fmt}]', colour='red', mininterval=0.2)
    else:
        pbar_energy.reset(total=initial_energy_val)
        pbar_energy.set_description(f"Convergence {n}x{n}")
        pbar_energy.n = 0 
        pbar_energy.last_print_n = 0
        pbar_energy.refresh()

    BATCH_SIZE = 5000 if n < 100 else 20000
    
    history_s = []
    history_e = []
    history_t = []
    history_b = []

    # --- Variables pour limiter la fréquence de rafraîchissement ---
    last_update_time = time.time()
    steps_accumulator = 0
    UPDATE_INTERVAL = 0.01

    while total_steps < steps_limit and current_energy > 0:
        
        this_batch = BATCH_SIZE
        if not infinite_mode and total_steps + this_batch > steps_limit:
            this_batch = steps_limit - total_steps
        
        current_energy, best_energy, temp, total_steps, n_acc, n_rej, n_imp, n_res = run_batch(
            current_board, block_coords_np, block_counts_np,
            current_energy, best_energy, best_board,
            row_counts, col_counts,
            temp, cooling_rate, initial_temp,
            this_batch, total_steps,
            n_accepted, n_rejected, n_improved, n_restarts
        )
        
        n_accepted, n_rejected, n_improved, n_restarts = n_acc, n_rej, n_imp, n_res
        
        history_s.append(total_steps)
        history_e.append(current_energy)
        history_t.append(temp)
        history_b.append(best_energy)
        
        # Accumuler les étapes
        steps_accumulator += this_batch
        
        # Vérifier si on doit mettre à jour l'affichage
        current_time = time.time()
        if (current_time - last_update_time > UPDATE_INTERVAL) or (current_energy == 0):
            pbar_iter.update(steps_accumulator)
            pbar_iter.set_postfix({"E": current_energy, "Best": best_energy, "T": f"{temp:.4f}", "R": n_restarts})
            
            if pbar_energy:
                progress = max(0, initial_energy_val - current_energy)
                pbar_energy.n = progress
                pbar_energy.refresh()
            
            steps_accumulator = 0
            last_update_time = current_time
        
    duration = time.time() - t_start
    
    # Mise à jour finale pour être sûr que le compte est bon
    if steps_accumulator > 0:
        pbar_iter.update(steps_accumulator)
    
    if own_pbars:
        pbar_iter.close()
        if pbar_energy: pbar_energy.close()
    
    if current_energy == 0 and pbar_energy:
        pbar_energy.n = initial_energy_val
        pbar_energy.refresh()

    return best_board, (history_s, history_e, history_t, history_b, duration, initial_energy_val, n_accepted, n_rejected, n_improved, n_restarts, temp, initial_temp, cooling_rate, total_steps, np.sum(fixed_mask))
```

```python
# Cellule 4 : Fonction de test
def test(csv_file, n_min_sqrt=2, n_max_sqrt=9, repetitions_par_n=50, remplissage_percentages=[10, 20, 30, 40, 50, 60, 70, 80, 90], infinite_mode=None):
    
    if not os.path.isfile(csv_file):
        with open(csv_file, mode='w', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(["N_Size", "Run_ID", "Fixed_Pct", "Success", "Duration", "Iterations", "Energy_Init", "Energy_Final", "Temp_Init", "Temp_Final", "Cooling_Rate", "Restarts", "Accepted", "Rejected", "Improved", "Fixed_Count"])
    
    display(HTML("<h3>Lancement de la campagne de tests intensifs...</h3>"))
    
    # --- PRÉPARATION DU MODÈLE DE PRÉDICTION ---
    tasks_metadata = []
    for n_sqrt in range(n_min_sqrt, n_max_sqrt + 1):
        n_size = n_sqrt**2
        for pct in remplissage_percentages:
            for _ in range(repetitions_par_n):
                tasks_metadata.append((n_size, pct))
    
    history_weights = {}
    if os.path.isfile(csv_file):
        try:
            df_hist = pd.read_csv(csv_file)
            stats = df_hist.groupby(['N_Size', 'Fixed_Pct'])['Duration'].mean()
            history_weights = stats.to_dict()
        except Exception:
            pass 
            
    task_weights = []
    for (n, pct) in tasks_metadata:
        if (n, pct) in history_weights:
            weight = history_weights[(n, pct)]
            weight = max(0.01, weight)
        else:
            scale_n = (n / 9.0) ** 4.5
            dist_peak = abs(pct - 40)
            if dist_peak < 15:
                difficulty = 1.0
            elif dist_peak < 30:
                difficulty = 0.2
            else:
                difficulty = 0.05
            weight = 0.2 * scale_n * difficulty
            weight = max(0.01, weight) 
            
        task_weights.append(weight)
        
    total_estimated_seconds = sum(task_weights)
    total_duration_str = format_duration(total_estimated_seconds)
    
    # --- INITIALISATION ---
    global_start_time = time.time()
    
    # Utilisation de mininterval pour éviter de surcharger le notebook
    pbar_global = tqdm(
        total=int(total_estimated_seconds), 
        desc=f"Progression (0s / {total_duration_str})", 
        position=0, 
        bar_format='{l_bar}{bar}| [ETA: {remaining}]',
        colour='green',
        mininterval=1.0
    )
    
    pbar_current_iter = tqdm(total=100, desc="Initialisation...", position=1, leave=True, colour='blue', mininterval=0.5)
    pbar_current_energy = tqdm(total=100, desc="Convergence", position=2, leave=True, bar_format='{l_bar}{bar}| {n_fmt}/{total_fmt}', colour='red', mininterval=0.5)
    
    task_idx = 0
    
    for n_sqrt in range(n_min_sqrt, n_max_sqrt + 1):
        n_size = n_sqrt**2
        
        for remplissage_pourcentage in remplissage_percentages:
            holes = int(n_size**2 * (1 - (remplissage_pourcentage / 100)))
            
            for i in range(repetitions_par_n):
                predicted_w = task_weights[task_idx]
                task_idx += 1
                
                sudoku = generate_sudoku(n_size, holes)
                
                use_infinite = infinite_mode
                if use_infinite is None:
                    use_infinite = (n_size > 81)
                
                solution, stats = solve_sudoku_annealing(
                    sudoku, 
                    infinite_mode=use_infinite, 
                    pbar_iter=pbar_current_iter, 
                    pbar_energy=pbar_current_energy
                )
                
                s, e, t, b, duration, e_init, n_acc, n_rej, n_imp, n_res, t_end, t_init, c_rate, steps, n_fixed = stats
                
                final_energy = calculate_energy(solution)
                success = (final_energy == 0)
                
                with open(csv_file, mode='a', newline='') as f:
                    writer = csv.writer(f)
                    writer.writerow([n_size, i+1, remplissage_pourcentage, success, duration, steps, e_init, final_energy, t_init, t_end, c_rate, n_res, n_acc, n_rej, n_imp, n_fixed])
                
                pbar_global.update(predicted_w)
                
                current_time_str = format_duration(pbar_global.n)
                pbar_global.set_description(f"Progression ({current_time_str} / {total_duration_str})")
                pbar_global.set_postfix({"Last": f"{n_size}x{n_size} ({remplissage_pourcentage}%)"})

    pbar_current_iter.close()
    pbar_current_energy.close()
    pbar_global.close()
    
    total_real_duration = time.time() - global_start_time
    
    display(HTML(f"<h3 style='color:green'>Campagne terminée. Toutes les statistiques sont enregistrées dans {csv_file}. Durée réelle : {format_duration(total_real_duration)}</h3>"))
```

```python
# Cellule 6 : Analyse Graphique
df = pd.read_csv("resultats_sudoku_complets.csv")

sns.set_theme(style="whitegrid", context="paper", font_scale=1.4)

MH_SIZE = 5
PLOT_CONFIG = {
    'fig_size': (12, 7),
    'palette': "flare",
    'cmap': "flare",
    'alpha': 0.75,
    'linewidth': 1,
    'edgecolor': 'none',
    'line_markersize': MH_SIZE,
    'scatter_s': MH_SIZE ** 2,
    'rotation': 270,
    'labelpad': 20,
    'pad': 20,
}

def put_legend_outside(ax, title="Légende", handles=None, labels=None):
    if handles and labels:
        ax.legend(handles=handles, labels=labels, title=title, bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0, frameon=True)
    else:
        ax.legend(title=title, bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0, frameon=True)

df['N_Label'] = df['N_Size'].apply(lambda x: f"$N=${x}")
n_sizes = sorted(df['N_Size'].unique())
hue_order = [f"$N=${n}" for n in n_sizes]
markers_list = ['o', '*', '^', 's', 'v', '<', '>', 'p', 'D']
markers_dict = {f"$N=${n}": markers_list[i % len(markers_list)] for i, n in enumerate(n_sizes)}

shape_handles = [Line2D([0], [0], marker=markers_dict[f"$N=${n}"], color='w', label=f"$N=${n}", markerfacecolor='#555555', markersize=8) for n in n_sizes]

plt.figure(figsize=PLOT_CONFIG['fig_size'])
ax = sns.lineplot(
    data=df, 
    x='Fixed_Pct', 
    y='Success', 
    hue='N_Label', 
    hue_order=hue_order,
    style='N_Label',
    style_order=hue_order,
    markers=markers_dict,
    dashes=False,
    linewidth=PLOT_CONFIG['linewidth'],
    palette=PLOT_CONFIG['palette'],
    markersize=PLOT_CONFIG['line_markersize'],
    err_style=None,
    markeredgecolor='none'
)
plt.title("Transition de phase de la résolubilité en fonction de la densité d'indices", pad=PLOT_CONFIG['pad'])
plt.xlabel("Taux de remplissage initial (%)")
plt.ylabel("Probabilité de succès ($P_{solve}$)")
plt.ylim(-0.05, 1.05)
put_legend_outside(plt.gca(), "Taille de Grille")
plt.tight_layout()
plt.show()

plt.figure(figsize=PLOT_CONFIG['fig_size'])
df_success = df[df['Success'] == True].copy()
df_success['Regime'] = pd.cut(df_success['Fixed_Pct'], bins=[0, 30, 60, 100], labels=['Difficile (<30%)', 'Moyen (30-60%)', 'Facile (>60%)'])
label_regime = {'Difficile (<30%)', 'Moyen (30-60%)', 'Facile (>60%)'}
for label in label_regime:
    subset = df_success[df_success['Regime'] == label]
    if not subset.empty:
        means = subset.groupby('N_Size')['Duration'].mean()
        plt.plot(means.index, means.values, marker='', linestyle='-', 
                 linewidth=PLOT_CONFIG['linewidth'], label=f"Tendance : {label}",
                 markersize=PLOT_CONFIG['line_markersize'])
ax = sns.scatterplot(
    data=df_success, 
    x='N_Size', 
    y='Duration', 
    hue='Fixed_Pct', 
    style='N_Label',
    markers=markers_dict,
    palette=PLOT_CONFIG['palette'], 
    alpha=PLOT_CONFIG['alpha'], 
    s=PLOT_CONFIG['scatter_s'],
    edgecolor='none',
    linewidth=0,
    legend=False
)
plt.yscale('log')
plt.xscale('log')
plt.title("Lois d'échelle temporelle selon le régime de contraintes (Log-Log)", pad=PLOT_CONFIG['pad'])
plt.xlabel("Taille de la Grille ($N$)")
plt.ylabel("Temps de calcul moyen (s)")
plt.xticks(n_sizes, labels=n_sizes)
plt.grid(True, which="minor", ls=":", alpha=0.4)
norm = plt.Normalize(df['Fixed_Pct'].min(), df['Fixed_Pct'].max())
sm = plt.cm.ScalarMappable(cmap=PLOT_CONFIG['cmap'], norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, pad=0.02)
cbar.set_label("Remplissage Initial (%)", rotation=PLOT_CONFIG['rotation'], labelpad=PLOT_CONFIG['labelpad'])
ax.legend(handles=shape_handles, title="Taille de Grille", bbox_to_anchor=(1.2, 1), loc='upper left')
plt.tight_layout()
plt.show()

plt.figure(figsize=PLOT_CONFIG['fig_size'])
ax = sns.lineplot(
    data=df, 
    x='Fixed_Pct', 
    y='Iterations', 
    hue='N_Label',
    hue_order=hue_order,
    style='N_Label',
    style_order=hue_order,
    markers=markers_dict,
    dashes=False,
    linewidth=PLOT_CONFIG['linewidth'],
    palette=PLOT_CONFIG['palette'],
    markersize=PLOT_CONFIG['line_markersize'],
    markeredgecolor='none'
)
plt.yscale('log')
plt.title("Phénomène de 'Peak Hardness' (Complexité en itérations)", pad=PLOT_CONFIG['pad'])
plt.xlabel("Taux de remplissage initial (%)")
plt.ylabel("Nombre d'itérations (Log)")
put_legend_outside(plt.gca(), "Taille de Grille")
plt.tight_layout()
plt.show()

plt.figure(figsize=PLOT_CONFIG['fig_size'])
df['N_Jitter'] = df['N_Size'] + np.random.normal(0, 0.15, len(df))
ax = sns.scatterplot(
    data=df,
    x='N_Jitter',
    y='Restarts',
    hue='Fixed_Pct',
    style='N_Label',
    markers=markers_dict,
    palette=PLOT_CONFIG['palette'],
    s=PLOT_CONFIG['scatter_s'],
    alpha=PLOT_CONFIG['alpha'],
    edgecolor='none',
    linewidth=0,
    legend=False
)
norm = plt.Normalize(df['Fixed_Pct'].min(), df['Fixed_Pct'].max())
sm = plt.cm.ScalarMappable(cmap=PLOT_CONFIG['cmap'], norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, pad=0.02)
cbar.set_label("Remplissage Initial (%)", rotation=PLOT_CONFIG['rotation'], labelpad=PLOT_CONFIG['labelpad'])
ax.legend(handles=shape_handles, title="Taille de Grille", bbox_to_anchor=(1.2, 1), loc='upper left')
plt.title("Analyse de la rugosité du paysage énergétique (Redémarrages)", pad=PLOT_CONFIG['pad'])
plt.xlabel("Taille de la Grille ($N$)")
plt.ylabel("Nombre de Redémarrages")
plt.xticks(n_sizes)
plt.grid(True, axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

plt.figure(figsize=PLOT_CONFIG['fig_size'])
ax = sns.scatterplot(
    data=df[df['Success'] == True],
    x='Energy_Init',
    y='Duration',
    hue='Fixed_Pct',
    style='N_Label',
    markers=markers_dict,
    palette=PLOT_CONFIG['palette'],
    s=PLOT_CONFIG['scatter_s'],
    alpha=PLOT_CONFIG['alpha'],
    edgecolor='none',
    linewidth=0,
    legend=False
)
norm = plt.Normalize(df['Fixed_Pct'].min(), df['Fixed_Pct'].max())
sm = plt.cm.ScalarMappable(cmap=PLOT_CONFIG['cmap'], norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, pad=0.02)
cbar.set_label("Remplissage Initial (%)", rotation=PLOT_CONFIG['rotation'], labelpad=PLOT_CONFIG['labelpad'])
ax.legend(handles=shape_handles, title="Taille de Grille", bbox_to_anchor=(1.2, 1), loc='upper left')
plt.title("Coût temporel en fonction de la distance énergétique initiale", pad=PLOT_CONFIG['pad'])
plt.xlabel("Énergie Initiale (Nombre de conflits à résoudre)")
plt.ylabel("Durée de Résolution (s)")
plt.tight_layout()
plt.show()
```
